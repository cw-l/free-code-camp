**Day 7: 2026 Winter Games Day 2: Snowboarding**

Given a snowboarder's starting stance and a rotation in degrees, determine their landing stance.

    A snowboarder's stance is either "Regular" or "Goofy".
    Trick rotations are multiples of 90 degrees. Positive indicates clockwise rotation, and negative indicate counter-clockwise rotation.
    The landing stance flips every 180 degrees of rotation.

For example, given "Regular" and 90, return "Regular". Given "Regular" and 180 degrees, return "Goofy".

Tests:

1. get_landing_stance("Regular", 90) should return "Regular".
2. get_landing_stance("Regular", 180) should return "Goofy".
3. get_landing_stance("Goofy", -270) should return "Regular".
4. get_landing_stance("Regular", 2340) should return "Goofy".

6. get_landing_stance("Goofy", -540) should return "Regular".
7. get_landing_stance("Goofy", 90) should return "Goofy".

```python
def get_landing_stance(start_stance, rotation):
    if abs(rotation) < 180: return start_stance
    else:
        if abs(rotation) // 180 % 2 == 0:
            return start_stance
        else:
            return "Goofy" if start_stance == "Regular" else "Regular"


assert get_landing_stance("Regular", 90) == "Regular", "Failed test case 1"
assert get_landing_stance("Regular", 180) == "Goofy", "Failed test case 2"
assert get_landing_stance("Goofy", -270) == "Regular", "Failed test case 3"
assert get_landing_stance("Regular", 2340) == "Goofy", "Failed test case 4"
assert get_landing_stance("Goofy", 2160) == "Goofy", "Failed test case 5"
assert get_landing_stance("Goofy", -540) == "Regular", "Failed test case 6"
print('PASS')
```

**Day 9: 2026 Winter Games Day 4: Ski Jumping**

Given distance points, style points, a wind compensation value, and K-point bonus value, calculate your score for the ski jump and determine if you won a medal or not.

    Your score is calculated by summing the above four values.

The current total scores of the other jumpers are:

165.5
172.0
158.0
180.0
169.5
175.0
162.0
170.0

    If your score is the best, return "Gold"
    If it's second best, return "Silver"
    If it's third best, return "Bronze"
    Otherwise, return "No Medal"

Tests:

1. ski_jump_medal(125.0, 58.0, 0.0, 6.0) should return "Gold".
2. ski_jump_medal(119.0, 50.0, 1.0, 4.0) should return "Bronze".
3. ski_jump_medal(122.0, 52.0, -1.0, 4.0) should return "Silver".
4. ski_jump_medal(118.0, 50.5, -1.5, 4.0) should return "No Medal".
5. ski_jump_medal(124.0, 50.5, 2.0, 5.0) should return "Gold".
6. ski_jump_medal(119.0, 49.5, 0.0, 3.0) should return "No Medal".

```python
scores = [ 165.5, 172.0, 158.0, 180.0, 169.5, 175.0, 162.0, 170.0 ]

def ski_jump_medal(distance_points, style_points, wind_comp, k_point_bonus):
    scores.sort(reverse=True)
    my_score = distance_points + style_points + wind_comp + k_point_bonus
    if my_score > scores[0]: return "Gold"
    elif my_score > scores[1]: return "Silver"
    elif my_score > scores[2]: return "Bronze"
    return "No Medal"


assert ski_jump_medal(125.0, 58.0, 0.0, 6.0) == "Gold", "Failed test case 1"
assert ski_jump_medal(119.0, 50.0, 1.0, 4.0) == "Bronze", "Failed test case 2"
assert ski_jump_medal(122.0, 52.0, -1.0, 4.0) == "Silver", "Failed test case 3"
assert ski_jump_medal(118.0, 50.5, -1.5, 4.0) == "No Medal", "Failed test case 4"
assert ski_jump_medal(124.0, 50.5, 2.0, 5.0) == "Gold", "Failed test case 5"
assert ski_jump_medal(119.0, 49.5, 0.0, 3.0) == "No Medal", "Failed test case 6"
print('PASS')
```

```python
def ski_jump_medal(distance_points, style_points, wind_comp, k_point_bonus):
    top_scores = sorted(scores, reverse=True)
    my_score = sum([distance_points, style_points, wind_comp, k_point_bonus])
    
    medals = zip(["Gold", "Silver", "Bronze"], top_scores)
    
    return next((medal for medal, cutoff in medals if my_score > cutoff), "No Medal")


assert ski_jump_medal(125.0, 58.0, 0.0, 6.0) == "Gold", "Failed test case 1"
assert ski_jump_medal(119.0, 50.0, 1.0, 4.0) == "Bronze", "Failed test case 2"
assert ski_jump_medal(122.0, 52.0, -1.0, 4.0) == "Silver", "Failed test case 3"
assert ski_jump_medal(118.0, 50.5, -1.5, 4.0) == "No Medal", "Failed test case 4"
assert ski_jump_medal(124.0, 50.5, 2.0, 5.0) == "Gold", "Failed test case 5"
assert ski_jump_medal(119.0, 49.5, 0.0, 3.0) == "No Medal", "Failed test case 6"
print('PASS')
```

Given an array of finish times for a cross-country ski race, convert them into times behind the winner.

    Given times are strings in "H:MM:SS" format.
    Given times will be in order from fastest to slowest.
    The winners time (fastest time) should correspond to "0".
    Each other time should show the time behind the winner, in the format "+M:SS".

For example, given ["1:25:32", "1:26:10", "1:27:05"], return ["0", "+0:38", "+1:33"].

Tests:

1. get_relative_results(["1:25:32", "1:26:10", "1:27:05"]) should return ["0", "+0:38", "+1:33"].
2. get_relative_results(["1:00:01", "1:00:05", "1:00:10"]) should return ["0", "+0:04", "+0:09"].
3. get_relative_results(["1:10:06", "1:10:23", "1:10:48", "1:12:11"]) should return ["0", "+0:17", "+0:42", "+2:05"].
4. get_relative_results(["0:49:13", "0:49:15", "0:50:14", "0:51:30", "0:51:58", "0:52:16", "0:53:12", "0:53:31", "0:56:19", "1:02:20"]) should return ["0", "+0:02", "+1:01", "+2:17", "+2:45", "+3:03", "+3:59", "+4:18", "+7:06", "+13:07"].
5. get_relative_results(["2:01:15", "2:10:45", "2:10:53", "2:11:04", "2:11:55", "2:13:27", "2:14:30", "2:15:10"]) should return ["0", "+9:30", "+9:38", "+9:49", "+10:40", "+12:12", "+13:15", "+13:55"].

```python
from datetime import datetime

def get_relative_results(results):
    dt_objs = sorted([datetime.strptime(t, "%H:%M:%S") for t in results])
    relative_results = [f"+{m}:{s:02d}" for t in dt_objs for m, s in [divmod(int((t - dt_objs[0]).total_seconds()), 60)]]
    relative_results[0] = '0'
    return relative_results


assert get_relative_results(["1:25:32", "1:26:10", "1:27:05"]) == ["0", "+0:38", "+1:33"], "Failed test case 1"
assert get_relative_results(["1:00:01", "1:00:05", "1:00:10"]) == ["0", "+0:04", "+0:09"], "Failed test case 2"
assert get_relative_results(["1:10:06", "1:10:23", "1:10:48", "1:12:11"]) == ["0", "+0:17", "+0:42", "+2:05"], "Failed test case 3"
assert get_relative_results(["0:49:13", "0:49:15", "0:50:14", "0:51:30", "0:51:58", "0:52:16", "0:53:12", "0:53:31", "0:56:19", "1:02:20"]) == ["0", "+0:02", "+1:01", "+2:17", "+2:45", "+3:03", "+3:59", "+4:18", "+7:06", "+13:07"], "Failed test case 4"
assert get_relative_results(["2:01:15", "2:10:45", "2:10:53", "2:11:04", "2:11:55", "2:13:27", "2:14:30", "2:15:10"]) == ["0", "+9:30", "+9:38", "+9:49", "+10:40", "+12:12", "+13:15", "+13:55"], "Failed test case 5"
print('PASS')
```

**Day 11: 2026 Winter Games Day 6: Figure Skating**

Given an array of judge scores and optional penalties, calculate the final score for a figure skating routine.

The first argument is an array of 10 judge scores, each a number from 0 to 10. Remove the highest and lowest judge scores and sum the remaining 8 scores to get the base score.

Any additional arguments passed to the function are penalties. Subtract all penalties from the base score to get the final score.

Tests:

1. compute_score([10, 8, 9, 6, 9, 8, 8, 9, 7, 7], 1) should return 64.
2. compute_score([10, 10, 10, 10, 10, 10, 10, 10, 10, 10]) should return 80.
3. compute_score([10, 8, 9, 10, 9, 8, 8, 9, 10, 7], 1, 2, 1) should return 67.
4. compute_score([8.0, 8.5, 9.0, 8.5, 9.0, 8.0, 9.0, 8.5, 9.0, 8.5], 0.5, 1.0) should return 67.5.
5. compute_score([6.0, 8.5, 7.0, 9.0, 7.5, 8.0, 6.5, 9.5, 7.0, 8.0], 1.5, 0.5, 0.5) should return 59.

```python
def compute_score(judge_scores, *penalties):
    return sum(sorted(judge_scores)[1:-1]) - sum(penalties)


assert compute_score([10, 8, 9, 6, 9, 8, 8, 9, 7, 7], 1) == 64, "Failed test case 1"
assert compute_score([10, 10, 10, 10, 10, 10, 10, 10, 10, 10]) == 80, "Failed test case 2"
assert compute_score([10, 8, 9, 10, 9, 8, 8, 9, 10, 7], 1, 2, 1) == 67, "Failed test case 3"
assert compute_score([8.0, 8.5, 9.0, 8.5, 9.0, 8.0, 9.0, 8.5, 9.0, 8.5], 0.5, 1.0) == 67.5, "Failed test case 4"
assert compute_score([6.0, 8.5, 7.0, 9.0, 7.5, 8.0, 6.5, 9.5, 7.0, 8.0], 1.5, 0.5, 0.5) == 59, "Failed test case 5"
print('PASS')
```

**Day 23: Blood Type Compatibility**

Given a donor blood type and a recipient blood type, determine whether the donor can give blood to the recipient.

Each blood type consists of:

    A letter: "A", "B", "AB", or "O"
    And an Rh factor: "+" or "-"

Blood types will be one of the valid letters followed by an Rh factor. For example, "AB+" and "O-" are valid blood types.

Letter Rules:

    "O" can donate to other letter type.
    "A" can donate to "A" and "AB".
    "B" can donate to "B" and "AB".
    "AB" can donate only to "AB".

Rh Rules:

    Negative ("-") can donate to both "-" and "+".
    Positive ("+") can donate only to "+".

Both letter and Rh rule must pass for a donor to be able to donate to the recipient.

Tests:
1. can_donate("B+", "B+") should return True.
2. can_donate("O-", "AB-") should return True.
3. can_donate("O+", "A-") should return False.
4. can_donate("A+", "AB+") should return True.
5. can_donate("A-", "B-") should return False.
6. can_donate("B-", "AB+") should return True.
7. can_donate("B-", "A+") should return False.
8. can_donate("O-", "O+") should return True.
9. can_donate("O+", "O-") should return False.
10. can_donate("AB+", "AB-") should return False.

```python
def can_donate(donor, recipient):
    if donor == recipient: return True

    if donor[0] == 'O' and donor[1] == '-':
        return True
    
    if donor[0] == 'O' and donor[1] == '+' and recipient[1] == '+':
        return True

    if donor[0] == 'A' and donor[1] == '-' and (recipient[0] == 'A' or recipient[0:2] == 'AB'):
        return True
     
    if donor[0] == 'A' and donor[1] == '+' and (recipient[0] == 'A' or recipient[0:2] == 'AB') and recipient[-1] == '+':
        return True

    if donor[0] == 'B' and donor[1] == '-' and (recipient[0] == 'B' or recipient[0:2] == 'AB'):
        return True
     
    if donor[0] == 'B' and donor[1] == '+' and (recipient[0] == 'B' or recipient[0:2] == 'AB') and recipient[-1] == '+':
        return True

    if donor[0:2] == 'AB' and donor[-1] == '-' and recipient[0:2] == 'AB':
        return True
     
    if donor[0:2] == 'AB' and donor[-1] == '+' and recipient[0:2] == 'AB' and recipient[-1] == '+':
        return True

    return False


assert can_donate("B+", "B+"), "Failed test case 1"
assert can_donate("O-", "AB-"), "Failed test case 2"
assert not can_donate("O+", "A-"), "Failed test case 3"
assert can_donate("A+", "AB+"), "Failed test case 4"
assert not can_donate("A-", "B-"), "Failed test case 5"
assert can_donate("B-", "AB+"), "Failed test case 6"
assert not can_donate("B-", "A+"), "Failed test case 7"
assert can_donate("O-", "O+"), "Failed test case 8"
assert not can_donate("O+", "O-"), "Failed test case 9"
assert not can_donate("AB+", "AB-"), "Failed test case 10"
print('PASS')
```

```python
def can_donate(donor, recipient):
    donor_type, donor_rh = donor[:-1], donor[-1]
    rec_type, rec_rh = recipient[:-1], recipient[-1]

    if donor_rh == '+' and rec_rh == '-':
        return False

    compatibility = {
        'O': {'O', 'A', 'B', 'AB'},
        'A': {'A', 'AB'},
        'B': {'B', 'AB'},
        'AB': {'AB'}
    }

    return rec_type in compatibility.get(donor_type, set())


assert can_donate("B+", "B+"), "Failed test case 1"
assert can_donate("O-", "AB-"), "Failed test case 2"
assert not can_donate("O+", "A-"), "Failed test case 3"
assert can_donate("A+", "AB+"), "Failed test case 4"
assert not can_donate("A-", "B-"), "Failed test case 5"
assert can_donate("B-", "AB+"), "Failed test case 6"
assert not can_donate("B-", "A+"), "Failed test case 7"
assert can_donate("O-", "O+"), "Failed test case 8"
assert not can_donate("O+", "O-"), "Failed test case 9"
assert not can_donate("AB+", "AB-"), "Failed test case 10"
print('PASS')
```

**Day 24: Business Day Count**

Given a start date and an end date, return the number of business days between the two.

    Given dates are in the format "YYYY-MM-DD".
    Weekdays are business days (Monday through Friday).
    Weekends are not business days (Saturday and Sunday).
    Include both the start and end dates when counting.

Tests:
1. count_business_days("2026-02-24", "2026-02-26") should return 3.
2. count_business_days("2026-02-24", "2026-02-28") should return 4.
3. count_business_days("2026-02-21", "2026-03-01") should return 5.
4. count_business_days("2026-03-08", "2026-03-17") should return 7.
5. count_business_days("2026-02-24", "2027-02-24") should return 262.

```python
from datetime import datetime, timedelta

def count_business_days(start, end):
    start_dt = datetime.strptime(start, "%Y-%m-%d")
    end_dt = datetime.strptime(end, "%Y-%m-%d")
    days_count = (end_dt - start_dt).days + 1

    return sum(1 for i in range(days_count) if (start_dt + timedelta(days=i)).weekday() < 5)


assert count_business_days("2026-02-24", "2026-02-26") == 3, "Failed test case 1"
assert count_business_days("2026-02-24", "2026-02-28") == 4, "Failed test case 2"
assert count_business_days("2026-02-21", "2026-03-01") == 5, "Failed test case 3"
assert count_business_days("2026-03-08", "2026-03-17") == 7, "Failed test case 4"
assert count_business_days("2026-02-24", "2027-02-24") == 262, "Failed test case 5"
print('PASS')
```

**Day 26: Letter and Number Count**

Given a string, return a message with the count of how many letters and numbers it contains.

    Letters are A-Z and a-z.
    Numbers are 0-9.
    Ignore all other characters.

Return "The string has X letters and Y numbers.", where "X" is the count of letters and "Y" is the count of numbers. If either count is 1, use the singular form for that item. E.g: "1 letter" instead of "1 letters" and "1 number" instead of "1 numbers".

Tests:
1. count_letters_and_numbers("helloworld123") should return "The string has 10 letters and 3 numbers.".
2. count_letters_and_numbers("Catch 22") should return "The string has 5 letters and 2 numbers.".
3. count_letters_and_numbers("A1!") should return "The string has 1 letter and 1 number.".
4. count_letters_and_numbers("12345") should return "The string has 0 letters and 5 numbers.".
5. count_letters_and_numbers("password") should return "The string has 8 letters and 0 numbers.".

```python
def count_letters_and_numbers(s):
    num_letters = 0
    num_numbers = 0
    for i in s:
        if i.isalpha(): num_letters += 1
        elif i.isdigit(): num_numbers += 1

    if num_letters == 1:
        letter_str = f"{num_letters} letter"
    elif num_letters > 1 or num_letters == 0:
        letter_str = f"{num_letters} letters"

    if num_numbers == 1:
        num_str = f"{num_numbers} number"
    elif num_numbers > 1 or num_numbers == 0:
        num_str = f"{num_numbers} numbers"

    return f"The string has {letter_str} and {num_str}."


assert count_letters_and_numbers("helloworld123") == "The string has 10 letters and 3 numbers.", "Failed test case 1."
assert count_letters_and_numbers("Catch 22") == "The string has 5 letters and 2 numbers.", "Failed test case 2."
assert count_letters_and_numbers("A1!") == "The string has 1 letter and 1 number.", "Failed test case 3."
assert count_letters_and_numbers("12345") == "The string has 0 letters and 5 numbers.", "Failed test case 4."
assert count_letters_and_numbers("password") == "The string has 8 letters and 0 numbers.", "Failed test case 5."
print('PASS')
```
