**Day 23: Blood Type Compatibility**

Given a donor blood type and a recipient blood type, determine whether the donor can give blood to the recipient.

Each blood type consists of:

    A letter: "A", "B", "AB", or "O"
    And an Rh factor: "+" or "-"

Blood types will be one of the valid letters followed by an Rh factor. For example, "AB+" and "O-" are valid blood types.

Letter Rules:

    "O" can donate to other letter type.
    "A" can donate to "A" and "AB".
    "B" can donate to "B" and "AB".
    "AB" can donate only to "AB".

Rh Rules:

    Negative ("-") can donate to both "-" and "+".
    Positive ("+") can donate only to "+".

Both letter and Rh rule must pass for a donor to be able to donate to the recipient.

Tests:
1. can_donate("B+", "B+") should return True.
2. can_donate("O-", "AB-") should return True.
3. can_donate("O+", "A-") should return False.
4. can_donate("A+", "AB+") should return True.
5. can_donate("A-", "B-") should return False.
6. can_donate("B-", "AB+") should return True.
7. can_donate("B-", "A+") should return False.
8. can_donate("O-", "O+") should return True.
9. can_donate("O+", "O-") should return False.
10. can_donate("AB+", "AB-") should return False.

```python
def can_donate(donor, recipient):
    if donor == recipient: return True

    if donor[0] == 'O' and donor[1] == '-':
        return True
    
    if donor[0] == 'O' and donor[1] == '+' and recipient[1] == '+':
        return True

    if donor[0] == 'A' and donor[1] == '-' and (recipient[0] == 'A' or recipient[0:2] == 'AB'):
        return True
     
    if donor[0] == 'A' and donor[1] == '+' and (recipient[0] == 'A' or recipient[0:2] == 'AB') and recipient[-1] == '+':
        return True

    if donor[0] == 'B' and donor[1] == '-' and (recipient[0] == 'B' or recipient[0:2] == 'AB'):
        return True
     
    if donor[0] == 'B' and donor[1] == '+' and (recipient[0] == 'B' or recipient[0:2] == 'AB') and recipient[-1] == '+':
        return True

    if donor[0:2] == 'AB' and donor[-1] == '-' and recipient[0:2] == 'AB':
        return True
     
    if donor[0:2] == 'AB' and donor[-1] == '+' and recipient[0:2] == 'AB' and recipient[-1] == '+':
        return True

    return False


assert can_donate("B+", "B+"), "Failed test case 1"
assert can_donate("O-", "AB-"), "Failed test case 2"
assert not can_donate("O+", "A-"), "Failed test case 3"
assert can_donate("A+", "AB+"), "Failed test case 4"
assert not can_donate("A-", "B-"), "Failed test case 5"
assert can_donate("B-", "AB+"), "Failed test case 6"
assert not can_donate("B-", "A+"), "Failed test case 7"
assert can_donate("O-", "O+"), "Failed test case 8"
assert not can_donate("O+", "O-"), "Failed test case 9"
assert not can_donate("AB+", "AB-"), "Failed test case 10"
print('PASS')
```

```python
def can_donate(donor, recipient):
    donor_type, donor_rh = donor[:-1], donor[-1]
    rec_type, rec_rh = recipient[:-1], recipient[-1]

    if donor_rh == '+' and rec_rh == '-':
        return False

    compatibility = {
        'O': {'O', 'A', 'B', 'AB'},
        'A': {'A', 'AB'},
        'B': {'B', 'AB'},
        'AB': {'AB'}
    }

    return rec_type in compatibility.get(donor_type, set())


assert can_donate("B+", "B+"), "Failed test case 1"
assert can_donate("O-", "AB-"), "Failed test case 2"
assert not can_donate("O+", "A-"), "Failed test case 3"
assert can_donate("A+", "AB+"), "Failed test case 4"
assert not can_donate("A-", "B-"), "Failed test case 5"
assert can_donate("B-", "AB+"), "Failed test case 6"
assert not can_donate("B-", "A+"), "Failed test case 7"
assert can_donate("O-", "O+"), "Failed test case 8"
assert not can_donate("O+", "O-"), "Failed test case 9"
assert not can_donate("AB+", "AB-"), "Failed test case 10"
print('PASS')
```

**Day 24: Business Day Count**

Given a start date and an end date, return the number of business days between the two.

    Given dates are in the format "YYYY-MM-DD".
    Weekdays are business days (Monday through Friday).
    Weekends are not business days (Saturday and Sunday).
    Include both the start and end dates when counting.

Tests:
1. count_business_days("2026-02-24", "2026-02-26") should return 3.
2. count_business_days("2026-02-24", "2026-02-28") should return 4.
3. count_business_days("2026-02-21", "2026-03-01") should return 5.
4. count_business_days("2026-03-08", "2026-03-17") should return 7.
5. count_business_days("2026-02-24", "2027-02-24") should return 262.

```python
from datetime import datetime, timedelta

def count_business_days(start, end):
    start_dt = datetime.strptime(start, "%Y-%m-%d")
    end_dt = datetime.strptime(end, "%Y-%m-%d")
    days_count = (end_dt - start_dt).days + 1

    return sum(1 for i in range(days_count) if (start_dt + timedelta(days=i)).weekday() < 5)


assert count_business_days("2026-02-24", "2026-02-26") == 3, "Failed test case 1"
assert count_business_days("2026-02-24", "2026-02-28") == 4, "Failed test case 2"
assert count_business_days("2026-02-21", "2026-03-01") == 5, "Failed test case 3"
assert count_business_days("2026-03-08", "2026-03-17") == 7, "Failed test case 4"
assert count_business_days("2026-02-24", "2027-02-24") == 262, "Failed test case 5"
print('PASS')
```

**Day 26: Letter and Number Count**

Given a string, return a message with the count of how many letters and numbers it contains.

    Letters are A-Z and a-z.
    Numbers are 0-9.
    Ignore all other characters.

Return "The string has X letters and Y numbers.", where "X" is the count of letters and "Y" is the count of numbers. If either count is 1, use the singular form for that item. E.g: "1 letter" instead of "1 letters" and "1 number" instead of "1 numbers".

Tests:
1. count_letters_and_numbers("helloworld123") should return "The string has 10 letters and 3 numbers.".
2. count_letters_and_numbers("Catch 22") should return "The string has 5 letters and 2 numbers.".
3. count_letters_and_numbers("A1!") should return "The string has 1 letter and 1 number.".
4. count_letters_and_numbers("12345") should return "The string has 0 letters and 5 numbers.".
5. count_letters_and_numbers("password") should return "The string has 8 letters and 0 numbers.".

```python
def count_letters_and_numbers(s):
    num_letters = 0
    num_numbers = 0
    for i in s:
        if i.isalpha(): num_letters += 1
        elif i.isdigit(): num_numbers += 1

    if num_letters == 1:
        letter_str = f"{num_letters} letter"
    elif num_letters > 1 or num_letters == 0:
        letter_str = f"{num_letters} letters"

    if num_numbers == 1:
        num_str = f"{num_numbers} number"
    elif num_numbers > 1 or num_numbers == 0:
        num_str = f"{num_numbers} numbers"

    return f"The string has {letter_str} and {num_str}."


assert count_letters_and_numbers("helloworld123") == "The string has 10 letters and 3 numbers.", "Failed test case 1."
assert count_letters_and_numbers("Catch 22") == "The string has 5 letters and 2 numbers.", "Failed test case 2."
assert count_letters_and_numbers("A1!") == "The string has 1 letter and 1 number.", "Failed test case 3."
assert count_letters_and_numbers("12345") == "The string has 0 letters and 5 numbers.", "Failed test case 4."
assert count_letters_and_numbers("password") == "The string has 8 letters and 0 numbers.", "Failed test case 5."
print('PASS')
```
