**Day 1: Schema Validator Part 1**

Given an object (JavaScript) or dictionary (Python), determine if it matches the following schema:

{
  username: string
}

    Extra keys are allowed

Tests:

1. is_valid_schema({"username": "bob"}) should return True.
2. is_valid_schema({"username": "jen", "posts": 30}) should return True.
3. is_valid_schema({"username": ""}) should return True.
4. is_valid_schema({"username": 7}) should return False.
5. is_valid_schema({"posts": 25}) should return False.

```python
def is_valid_schema(obj):
    if type(obj) is not dict:
        return False
        
    if type(obj.get("username")) is not str:
        return False

    return True


assert is_valid_schema({"username": "bob"}), "Failed test case 1"
assert is_valid_schema({"username": "jen", "posts": 30}), "Failed test case 2"
assert is_valid_schema({"username": ""}), "Failed test case 3"
assert not is_valid_schema({"username": 7}), "Failed test case 4"
assert not is_valid_schema({"posts": 25}), "Failed test case 5"
print('PASS') 
```

```python
import jsonschema
from jsonschema import validate

def is_valid_schema(obj):
    user_schema = {
        "type": "object",
        "properties": {
            "username": {"type": "string"},
        },
        "required": ["username"]
    }
    try:
        validate(instance=obj, schema=user_schema)
        return True
    except jsonschema.exceptions.ValidationError:
        return False


assert is_valid_schema({"username": "bob"}), "Failed test case 1"
assert is_valid_schema({"username": "jen", "posts": 30}), "Failed test case 2"
assert is_valid_schema({"username": ""}), "Failed test case 3"
assert not is_valid_schema({"username": 7}), "Failed test case 4"
assert not is_valid_schema({"posts": 25}), "Failed test case 5"
print('PASS')
```

  username: string,
  posts: number,
  verified: boolean,
  role: Roles,
  supporter?: boolean,
  badges: string[]
}

{
  users: UserProfile[]
}

    The pipe (|) symbol means "or". role must be one of the listed Roles values.
    The question mark (?) after supporter means that the field is optional, but is the specified type if it exists.
    UserProfile[] denotes an array of UserProfile objects. An empty array is valid.
    Extra keys are allowed

Tests:

1. is_valid_schema({"users": [{"username": "ron", "posts": 14, "verified": True, "role": "creator", "badges": ["early-adopter"]}, {"username": "cher", "posts": 25, "verified": True, "role": "moderator", "supporter": True, "followers": 20, "badges": ["helper"]}]}) should return True.

**Day 11: Idea Rankings**

Given a 2D array where each inner array contains (in this order) an idea name, an optimistic estimate, a realistic estimate, and a pessimistic estimate (in days), return an array of the idea names sorted by expected time to completion, shortest first.

Calculate the expected time to completion for each idea using the following formula:

    expected = ((optimistic + 4 * realistic + pessimistic) / 6) * length of idea name

Tests:

1. analyze_ideas([["Add logging", 2, 5, 15], ["SEO optimization", 4, 8, 20], ["Fix bug", 1, 3, 5]]) should return ["Fix bug", "Add logging", "SEO optimization"].
2. analyze_ideas([["Dark mode", 1, 3, 8], ["Real-time collaboration feature", 6, 12, 20], ["Add tooltip", 1, 2, 4]]) should return ["Add tooltip", "Dark mode", "Real-time collaboration feature"].
3. analyze_ideas([["Update user profile page", 3, 7, 14], ["Add pagination", 2, 5, 10], ["Add tags", 2, 3, 6], ["Fix login bug", 1, 4, 8]]) should return ["Add tags", "Fix login bug", "Add pagination", "Update user profile page"].
4. analyze_ideas([["Migrate database", 14, 25, 40], ["Add chat assistant", 8, 15, 24], ["Redesign onboarding flow", 3, 7, 13], ["Add language support", 6, 11, 18]]) should return ["Redesign onboarding flow", "Add language support", "Add chat assistant", "Migrate database"].
5. analyze_ideas([["Add email notifications", 3, 7, 10], ["Migrate deployment flow", 6, 10, 16], ["Add push notifications", 2, 6, 10], ["Optimize continuous integration", 5, 8, 15], ["Analyze user patterns", 5, 10, 18], ["Create onboarding curriculum", 6, 15, 25]]) should return ["Add push notifications", "Add email notifications", "Analyze user patterns", "Migrate deployment flow", "Optimize continuous integration", "Create onboarding curriculum"].

```python
def analyze_ideas(ideas):
    comp = []
    for idea in ideas:
        expected = ((idea[1] + 4 * idea[2] + idea[3]) / 6) * len(idea[0])
        comp.append([idea[0], expected])

    return [x[0] for x in sorted(comp, key=lambda x: x[1])]


assert analyze_ideas([["Add logging", 2, 5, 15], ["SEO optimization", 4, 8, 20], ["Fix bug", 1, 3, 5]]) == ["Fix bug", "Add logging", "SEO optimization"], "Failed test case 1"
assert analyze_ideas([["Dark mode", 1, 3, 8], ["Real-time collaboration feature", 6, 12, 20], ["Add tooltip", 1, 2, 4]]) == ["Add tooltip", "Dark mode", "Real-time collaboration feature"], "Failed test case 2"
assert analyze_ideas([["Update user profile page", 3, 7, 14], ["Add pagination", 2, 5, 10], ["Add tags", 2, 3, 6], ["Fix login bug", 1, 4, 8]]) == ["Add tags", "Fix login bug", "Add pagination", "Update user profile page"], "Failed test case 3"
assert analyze_ideas([["Migrate database", 14, 25, 40], ["Add chat assistant", 8, 15, 24], ["Redesign onboarding flow", 3, 7, 13], ["Add language support", 6, 11, 18]]) == ["Redesign onboarding flow", "Add language support", "Add chat assistant", "Migrate database"], "Failed test case 4"
assert analyze_ideas([["Add email notifications", 3, 7, 10], ["Migrate deployment flow", 6, 10, 16], ["Add push notifications", 2, 6, 10], ["Optimize continuous integration", 5, 8, 15], ["Analyze user patterns", 5, 10, 18], ["Create onboarding curriculum", 6, 15, 25]]) == ["Add push notifications", "Add email notifications", "Analyze user patterns", "Migrate deployment flow", "Optimize continuous integration", "Create onboarding curriculum"], "Failed test case 5"
print('PASS')
```

**Day 19: Rental Cost**

Given a rental timestamp, a return timestamp, and a rental tier, return the total cost of the rental including any late fees.

    Given timestamps are UTC ISO strings, for example: "2026-06-18T18:30:00Z".

    Rentals are due back by 12:00 PM UTC or earlier on the last day of the rental period. For example, a 1-day rental checked out at any time on March 15 is due back by 12:00 PM UTC on March 16.
    Each day past the due date and time incurs a late fee.

Pricing is as follows:
Tier 	Base cost 	Late fee per day
1 day 	$4.99 	$3.99
3 days 	$3.99 	$2.99
7 days 	$2.99 	$0.99

Return the total cost rounded to two decimal places in the format "$D.CC".

Tests:

1. get_rental_cost("2026-06-18T18:30:00Z", "2026-06-19T10:30:00Z", 1) should return "$4.99".
2. get_rental_cost("2026-06-18T14:30:00Z", "2026-06-20T12:30:00Z", 1) should return "$12.97".

```python
from datetime import datetime, timedelta
import math

cost_table = {
    "1": { "base": 4.99, "late": 3.99 },
    "3": { "base": 3.99, "late": 2.99 },
    "7": { "base": 2.99, "late": 0.99 }
}

def get_rental_cost(rented, returned, tier):
    rented_dt = datetime.fromisoformat(rented)

    due_date = rented_dt + timedelta(days=tier)

    # set due date to be 12:00:00.000000 UTC
    due_date = due_date.replace(hour=12, minute=0, second=0, microsecond=0)

    returned_dt = datetime.fromisoformat(returned)

    if returned_dt <= due_date:
        return f"${cost_table[str(tier)]['base']:.2f}"
    else:
        delta = returned_dt - due_date
        days_overdue = math.ceil(delta.total_seconds() / 86400)
        return f"${cost_table[str(tier)]['base'] + cost_table[str(tier)]['late'] * days_overdue:.2f}"


assert get_rental_cost("2026-06-18T18:30:00Z", "2026-06-19T10:30:00Z", 1) == "$4.99", "Failed test case 1"
assert get_rental_cost("2026-06-18T14:30:00Z", "2026-06-20T12:30:00Z", 1) == "$12.97", "Failed test case 2"
assert get_rental_cost("2026-06-18T10:15:00Z", "2026-06-18T19:45:00Z", 3) == "$3.99", "Failed test case 3"
assert get_rental_cost("2026-06-18T15:20:00Z", "2026-06-23T08:10:00Z", 3) == "$9.97", "Failed test case 4"
assert get_rental_cost("2026-06-18T12:00:00Z", "2026-06-25T12:00:00Z", 7) == "$2.99", "Failed test case 5"
assert get_rental_cost("2026-06-18T08:00:00Z", "2027-06-18T14:00:00Z", 7) == "$358.40", "Failed test case 6"
print('PASS')
```

```python
import math
from datetime import datetime, timedelta

COST_TABLE = {
    1: {"base": 4.99, "late": 3.99},
    3: {"base": 3.99, "late": 2.99},
    7: {"base": 2.99, "late": 0.99}
}

def get_rental_cost(rented: str, returned: str, tier: int) -> str:
    rates = COST_TABLE[tier]
    
    rented_dt = datetime.fromisoformat(rented)
    returned_dt = datetime.fromisoformat(returned)
    
    due_date = (rented_dt + timedelta(days=tier)).replace(hour=12, minute=0, second=0, microsecond=0)
    
    delta_seconds = (returned_dt - due_date).total_seconds()
    days_overdue = math.ceil(max(0.0, delta_seconds) / 86400)
    
    total_cost = rates["base"] + (rates["late"] * days_overdue)
    return f"${total_cost:.2f}"


assert get_rental_cost("2026-06-18T18:30:00Z", "2026-06-19T10:30:00Z", 1) == "$4.99", "Failed test case 1"
assert get_rental_cost("2026-06-18T14:30:00Z", "2026-06-20T12:30:00Z", 1) == "$12.97", "Failed test case 2"
assert get_rental_cost("2026-06-18T10:15:00Z", "2026-06-18T19:45:00Z", 3) == "$3.99", "Failed test case 3"
assert get_rental_cost("2026-06-18T15:20:00Z", "2026-06-23T08:10:00Z", 3) == "$9.97", "Failed test case 4"
assert get_rental_cost("2026-06-18T12:00:00Z", "2026-06-25T12:00:00Z", 7) == "$2.99", "Failed test case 5"
assert get_rental_cost("2026-06-18T08:00:00Z", "2027-06-18T14:00:00Z", 7) == "$358.40", "Failed test case 6"
print('PASS')
```

**Day 20: Prime Factorization**

Given an integer greater than 1, return its prime factorization as an array of numbers in ascending order.

A prime factorization is the set of prime numbers that multiply together to produce the given integer. Each number has exactly one set. For example, the prime factorization of 20 is [2, 2, 5] because 2 * 2 * 5 = 20.

If the given integer is itself prime, return it in a single-element array.

Tests:

1. prime_factorization(20) should return [2, 2, 5].
2. prime_factorization(17) should return [17].
3. prime_factorization(15) should return [3, 5].
4. prime_factorization(35) should return [5, 7].
5. prime_factorization(999) should return [3, 3, 3, 37].
6. prime_factorization(360) should return [2, 2, 2, 3, 3, 5].
7. prime_factorization(510510) should return [2, 3, 5, 7, 11, 13, 17].

```python
def prime_factorization(n):
    factors = []
    factor = 2
    
    while factor * factor <= n:
        while n % factor == 0:
            factors.append(factor)
            n //= factor
        factor += 1

    if n > 1:
        factors.append(n)
        
    return factors


assert prime_factorization(20) == [2, 2, 5], "Failed test case 1"
assert prime_factorization(17) == [17], "Failed test case 2"
assert prime_factorization(15) == [3, 5], "Failed test case 3"
assert prime_factorization(35) == [5, 7], "Failed test case 4"
assert prime_factorization(999) == [3, 3, 3, 37], "Failed test case 5"
assert prime_factorization(360) == [2, 2, 2, 3, 3, 5], "Failed test case 6"
assert prime_factorization(510510) == [2, 3, 5, 7, 11, 13, 17], "Failed test case 7"
print('PASS')
```
