**Day 1: Anagram Groups**

Given an array of words, return a 2d array of the words grouped into anagrams.

    Words are anagrams if they contain the same letters in any order.
    Each word belongs to exactly one group.
    Return order doesn't matter.

For example, given ["listen", "silent", "hello", "enlist", "world"], return [["listen", "silent", "enlist"], ["hello"], ["world"]].

*Tests*
1. group_anagrams(["listen", "silent", "hello", "enlist", "world"]) should return [["listen", "silent", "enlist"], ["hello"], ["world"]].
2. group_anagrams(["eat", "tea", "tan", "ate", "nat", "bat"]) should return [["ate", "eat", "tea"], ["bat"], ["nat", "tan"]].
3. group_anagrams(["care", "race", "acre", "pots", "stop", "tops", "opts", "post", "spot", "evil", "vile", "live", "veil"]) should return [["acre", "care", "race"], ["evil", "live", "veil", "vile"], ["opts", "post", "pots", "spot", "stop", "tops"]].
4. group_anagrams(["algorithms", "logarithms", "education", "cautioned", "auctioned", "triangle", "integral", "alerting", "relating"]) should return [["alerting", "integral", "relating", "triangle"], ["algorithms", "logarithms"], ["auctioned", "cautioned", "education"]].

```python
from collections import defaultdict

def group_anagrams(words):
    anagram_map = defaultdict(list)
    
    for word in words:
        sorted_key = ''.join(sorted(word))
        anagram_map[sorted_key].append(word)
        
    return list(anagram_map.values())


# Helper function to normalize the output for reliable comparison
def normalize(result):
    # Sort the words inside each group, then sort the groups themselves
    return sorted([sorted(group) for group in result])

# --- Test Case 1 ---
result1 = group_anagrams(["listen", "silent", "hello", "enlist", "world"])
expected1 = [["listen", "silent", "enlist"], ["hello"], ["world"]]
assert normalize(result1) == normalize(expected1), "Failed Case 1"

# --- Test Case 2 ---
result2 = group_anagrams(["eat", "tea", "tan", "ate", "nat", "bat"])
expected2 = [["ate", "eat", "tea"], ["bat"], ["nat", "tan"]]
assert normalize(result2) == normalize(expected2), "Failed Case 2"

# --- Test Case 3 ---
result3 = group_anagrams(["care", "race", "acre", "pots", "stop", "tops", "opts", "post", "spot", "evil", "vile", "live", "veil"])
expected3 = [["acre", "care", "race"], ["evil", "live", "veil", "vile"], ["opts", "post", "pots", "spot", "stop", "tops"]]
assert normalize(result3) == normalize(expected3), "Failed Case 3"

# --- Test Case 4 ---
result4 = group_anagrams(["algorithms", "logarithms", "education", "cautioned", "auctioned", "triangle", "integral", "alerting", "relating"])
expected4 = [["alerting", "integral", "relating", "triangle"], ["algorithms", "logarithms"], ["auctioned", "cautioned", "education"]]
assert normalize(result4) == normalize(expected4), "Failed Case 4"

print('PASS')
```

**Day 3: Good Day**

Given a time string in "HH:MM" format (24-hour clock), return:

    "Good morning" for times 05:00 to 11:59
    "Good afternoon" for times 12:00 to 17:59
    "Good evening" for times 18:00 to 21:59
    "Good night" for times 22:00 to 04:59

*Tests*
1. get_greeting("06:30") should return "Good morning".
2. get_greeting("12:00") should return "Good afternoon".
3. get_greeting("21:59") should return "Good evening".
4. get_greeting("00:01") should return "Good night".
5. get_greeting("11:30") should return "Good morning".

```python
from datetime import datetime

def get_greeting(s):
    morning = datetime.strptime('11:59','%H:%M').time()
    afternoon = datetime.strptime('17:59','%H:%M').time()
    evening = datetime.strptime('21:59','%H:%M').time()
    night = datetime.strptime('04:59','%H:%M').time()

    time = datetime.strptime(s,'%H:%M').time()

    if time <= night:
        s = 'Good night'
    elif time <= morning and time > night:
        s = 'Good morning'
    elif time <= afternoon:
        s = 'Good afternoon'
    elif time <= evening:
        s = 'Good evening'

    return s


assert get_greeting("06:30") == "Good morning", "Failed Case 1"
assert get_greeting("12:00") == "Good afternoon", "Failed Case 2"
assert get_greeting("21:59") == "Good evening", "Failed Case 3"
assert get_greeting("00:01") == "Good night", "Failed Case 4"
assert get_greeting("11:30") == "Good morning", "Failed Case 5"
print('PASS')
```

**Day 5: Narcissistic Number**

Given a positive integer, determine whether it is a narcissistic number.

* A number is narcissistic if the sum of each of its digits raised to the power of the total number of digits equals the number itself.

For example, 153 has 3 digits, and 13 + 53 + 33 = 153, so it is narcissistic.

*Tests*
1. is_narcissistic(153) should return True.
2. is_narcissistic(154) should return False.
3. is_narcissistic(371) should return True.
4. is_narcissistic(512) should return False.
5. is_narcissistic(9) should return True.
6. is_narcissistic(11) should return False.
7. is_narcissistic(9474) should return True.
8. is_narcissistic(6549) should return False.

```python
def is_narcissistic(n):
    digits = str(n)
    power = len(digits)
    total = 0

    for digit in digits:
        total += int(digit) ** power

    if total == n:
        return True

    return False


assert is_narcissistic(153) == True, "Failed Case 1"
assert is_narcissistic(154) == False, "Failed Case 2"
assert is_narcissistic(371) == True, "Failed Case 3"
assert is_narcissistic(512) == False, "Failed Case 4"
assert is_narcissistic(9) == True, "Failed Case 5"
assert is_narcissistic(11) == False, "Failed Case 6"
assert is_narcissistic(9474) == True, "Failed Case 7"
assert is_narcissistic(6549) == False, "Failed Case 8"
print("PASS")
```

* Each meal is in the format [meal, allergens], where meal is the name of the meal, and allergens is an array of the allergens the meal contains. For example, ["pasta", ["wheat", "milk"]].
* Allergens to avoid will be an array of strings.

Return safe meal names in the same order given. If no meal is safe, return an empty array.

*Tests*
1. get_allergen_friendly_meals([["pasta", ["wheat", "milk"]], ["salad", ["nuts"]]], ["milk"]) should return ["salad"].
2. get_allergen_friendly_meals([["steak", ["soy"]], ["fried rice", []], ["fish tacos", ["fish", "wheat"]], ["chicken parmesan", ["wheat", "milk"]]], ["soy", "fish"]) should return ["fried rice", "chicken parmesan"].
3. get_allergen_friendly_meals([["oatmeal", ["nuts"]], ["pancakes", ["wheat", "milk"]], ["granola", []], ["yogurt", ["milk"]], ["eggs", ["eggs", "milk"]], ["toast", ["wheat"]]], ["eggs", "milk"]) should return ["oatmeal", "granola", "toast"].
4. get_allergen_friendly_meals([["oatmeal", ["nuts"]], ["pancakes", ["wheat", "milk"]], ["granola", []], ["yogurt", ["milk"]], ["eggs", ["eggs", "milk"]], ["toast", ["wheat"]]], ["wheat", "nuts"]) should return ["granola", "yogurt", "eggs"].

Given a matrix (an array of arrays), return the transposed version of it.

To transpose the matrix, swap the rows and columns. E.g: a value at index [0, 1] should move to index [1, 0].

For example, given:
[
  [1, 2, 3],
  [4, 5, 6]
]

Return:

[
  [1, 4],
  [2, 5],
  [3, 6]
]

*Tests*
1. transpose([[1, 2, 3], [4, 5, 6]]) should return [[1, 4], [2, 5], [3, 6]].
2. transpose([[1, 2], [3, 4], [5, 6]]) should return [[1, 3, 5], [2, 4, 6]].
3. transpose([[1, 2], [3, 4], [5, 6], [7, 8]]) should return [[1, 3, 5, 7], [2, 4, 6, 8]].
4. transpose([["a", "b", "c"], ["d", "e", "f"], ["g", "h", "i"], ["j", "k", "l"]]) should return [["a", "d", "g", "j"], ["b", "e", "h", "k"], ["c", "f", "i", "l"]].

**Day 10: ISBN-13 Validator**

Given a string, determine if it is a valid ISBN-13 number.

A valid ISBN-13:

* Contains only digits and hyphens
* Has exactly 13 digits after removing hyphens
* Passes the following check:
  * Multiply each digit by 1 or 3, alternating (multiply the first digit by 1, the second by 3, the third by 1, and so on).
  * The sum of the results must be divisible by 10.

*Tests*
1. is_valid_isbn_13("9780306406157") should return True.
2. is_valid_isbn_13("97803064061570") should return False.
3. is_valid_isbn_13("978-0-13-595705-9") should return True.
4. is_valid_isbn_13("978-030-64061A-4") should return False.
5. is_valid_isbn_13("9-7-8-0-1-3-4-7-5-7-5-9-9") should return True.

```python
import re

def is_valid_isbn_13(s):
    pattern = r"^[0-9-]+$"
    if re.match(pattern, s):
        digits = s.replace("-", "")
        if len(digits) != 13:
            return False
        else:
            sum = 0
            for i in range(13):
                if i % 2 == 0:
                    sum += int(digits[i])
                else:
                    sum += int(digits[i]) * 3
            if sum % 10 != 0:
                return False
    else:
        return False
    return True


assert is_valid_isbn_13("9780306406157") == True, "Failed Case 1"
assert is_valid_isbn_13("97803064061570") == False, "Failed Case 2"
assert is_valid_isbn_13("978-0-13-595705-9") == True, "Failed Case 3"
assert is_valid_isbn_13("978-030-64061A-4") == False, "Failed Case 4"
assert is_valid_isbn_13("9-7-8-0-1-3-4-7-5-7-5-9-9") == True, "Failed Case 5"
print("PASS")
```

4. get_oldest([{ "name": "George", "age": 50 }, { "name": "Shirley", "age": 42 }, { "name": "Beth", "age": 48 }, { "name": "Holly", "age": 50 }, { "name": "Kevin", "age": 44 }, { "name": "Frank", "age": 47 }, { "name": "Zach", "age": 50 }, { "name": "Jennifer", "age": 43 }]) should return ["George", "Holly", "Zach"].

**Day 13: Offending Element**

Given an array of integers that is sorted in ascending order except for one out-of-place element, return the index of that element.

    If more than one element could be considered out of place, return the index of the first one.

*Tests*
1. find_offender([1, 6, 2, 3, 4, 5]) should return 1.
2. find_offender([1, 2, 3, 5, 4, 5]) should return 3.
3. find_offender([2, 1]) should return 0.
4. find_offender([2, 4, 1, 6, 8]) should return 2.
5. find_offender([5, 18, 24, 33, 40, 55, 15, 68, 84, 91]) should return 6.

```python
def find_offender(arr):
    for i in range(len(arr)-1):
        if arr[i] >= arr[i+1]: 
            if i == 0 or arr[i-1] < arr[i+1]:
                return i
            return i+1
    return None


assert find_offender([1, 6, 2, 3, 4, 5]) == 1, "Failed Case 1"
assert find_offender([1, 2, 3, 5, 4, 5]) == 3, "Failed Case 2"
assert find_offender([2, 1]) == 0, "Failed Case 3"
assert find_offender([2, 4, 1, 6, 8]) == 2, "Failed Case 4"
assert find_offender([5, 18, 24, 33, 40, 55, 15, 68, 84, 91]) == 6, "Failed Case 5"
print("PASS")
```

7. is_mirror_image("><(((*>", "<*)))><") should return True.
8. is_mirror_image("WTYUIOHAXVMwoxv08=+:|-_*^!.[]{}<>bdpq()", "()pqbd<>{}[].!^*_-|:+=80vxowMVXAHOIUYTW") should return True.

```python
def is_mirror_image(s1, s2):
    # 1. Mirroring requires identical lengths
    if len(s1) != len(s2):
        return False

    # 2. Define the transformation rules
    mirror_map = {
        # Symmetric (Self-mirroring)
        'W':'W', 'T':'T', 'Y':'Y', 'U':'U', 'I':'I', 'O':'O', 'H':'H', 
        'A':'A', 'X':'X', 'V':'V', 'M':'M', 'w':'w', 'o':'o', 'x':'x', 
        'v':'v', '0':'0', '8':'8', '=':'=', '+':'+', ':':':', '|':'|', 
        '-':'-', '_':'_', '*':'*', '^':'^', '!':'!', '.':'.', ' ':' ',
        # Pairs (Mutual mirroring)
        '[':']', ']':'[', 
        '{':'}', '}':'{', 
        '<':'>', '>':'<', 
        '(':')', ')':'(',
        'b':'d', 'd':'b', 
        'p':'q', 'q':'p'
    }

    # 3. Compare s1 to the reversed s2 (using s2[::-1])
    # For every pair, the character in s1 must be the mirror of the char in s2
    for c1, c2 in zip(s1, s2[::-1]):
        if mirror_map.get(c1) != c2:
            return False
            
    return True


assert is_mirror_image("[HOW]", "[WOH]") == True, "Failed Case 1"
assert is_mirror_image("MOM", "MOM") == True, "Failed Case 2"
assert is_mirror_image("vow", "wov") == True, "Failed Case 3"
assert is_mirror_image("TIM", "TIM") == False, "Failed Case 4"
assert is_mirror_image("{WOW}", "}WOW{") == False, "Failed Case 5"
assert is_mirror_image("XXVII", "IIV%X") == False, "Failed Case 6"
assert is_mirror_image("><(((*>", "<*)))><") == True, "Failed Case 7"
assert is_mirror_image("WTYUIOHAXVMwoxv08=+:|-_*^!.[]{}<>bdpq()", "()pqbd<>{}[].!^*_-|:+=80vxowMVXAHOIUYTW") == True, "Failed Case 8"
print("PASS")
```

**Day 17: Mongo ID Date**

Given a MongoDB ID string, return its creation time as an ISO 8601 string.
* A MongoDB ID is a 24-character hex string. The first 8 characters represent a Unix timestamp (in seconds) encoded as a base-16 integer.

For example, "6a094b50bcf86cd799439011" has a timestamp of "6a094b50" in hex, which is 1778994000 in decimal, representing a creation time of "2026-05-17T05:00:00.000Z".

*Tests*
1. mongo_id_to_date("6a094b50bcf86cd799439011") should return "2026-05-17T05:00:00.000Z".
2. mongo_id_to_date("695344eb1f4a4c1123042128") should return "2025-12-30T03:20:11.000Z".
3. mongo_id_to_date("386da62df34123ac54617e56") should return "2000-01-01T07:01:01.000Z".
4. mongo_id_to_date("69f571c3d7711807afd3dd55") should return "2026-05-02T03:38:43.000Z".
5. mongo_id_to_date("68adce01c0e1144d0a90295a") should return "2025-08-26T15:08:49.000Z".

```python
import string
from datetime import datetime, timezone

def ismongodbid(s):
    return len(s) == 24 and all(c in string.hexdigits for c in s)

def mongo_id_to_date(s):
    if ismongodbid(s):
        dt = datetime.fromtimestamp(int(s[:8], 16), tz=timezone.utc)
        iso_string = dt.strftime("%Y-%m-%dT%H:%M:%S.%f")[:23] + "Z"
        return iso_string  
    return "None"


assert mongo_id_to_date("6a094b50bcf86cd799439011") == "2026-05-17T05:00:00.000Z", "Failed Case 1"
assert mongo_id_to_date("695344eb1f4a4c1123042128") == "2025-12-30T03:20:11.000Z", "Failed Case 2"
assert mongo_id_to_date("386da62df34123ac54617e56") == "2000-01-01T07:01:01.000Z", "Failed Case 3"
assert mongo_id_to_date("69f571c3d7711807afd3dd55") == "2026-05-02T03:38:43.000Z", "Failed Case 4"
assert mongo_id_to_date("68adce01c0e1144d0a90295a") == "2025-08-26T15:08:49.000Z", "Failed Case 5"
print("PASS")
```

```python
import re
from datetime import datetime, timezone

# Pre-compile the regex pattern for maximum performance
# Matches exactly 24 lowercase hexadecimal characters
MONGO_ID_PATTERN = re.compile(r"^[0-9a-f]{24}$")

def is_mongodb_id(s):
    """Validates if a string is a valid MongoDB ObjectId."""
    # Using isinstance protects the regex from throwing an error if passed an integer/None
    return bool(isinstance(s, str) and MONGO_ID_PATTERN.match(s))

def mongo_id_to_date(s):
    """Extracts the creation date from a MongoDB ObjectId as an ISO string."""
    if not is_mongodb_id(s):
        return "None"  # Or consider returning None (the object) instead of the string "None"
        
    # Extract first 4 bytes (8 hex chars) and convert from base-16 timestamp
    timestamp = int(s[:8], 16)
    dt = datetime.fromtimestamp(timestamp, tz=timezone.utc)
    
    # Formats directly to 'YYYY-MM-DDTHH:MM:SS.mmmZ' safely without manual string slicing
    return dt.isoformat(timespec="milliseconds").replace("+00:00", "Z")


assert mongo_id_to_date("6a094b50bcf86cd799439011") == "2026-05-17T05:00:00.000Z", "Failed Case 1"
assert mongo_id_to_date("695344eb1f4a4c1123042128") == "2025-12-30T03:20:11.000Z", "Failed Case 2"
assert mongo_id_to_date("386da62df34123ac54617e56") == "2000-01-01T07:01:01.000Z", "Failed Case 3"
assert mongo_id_to_date("69f571c3d7711807afd3dd55") == "2026-05-02T03:38:43.000Z", "Failed Case 4"
assert mongo_id_to_date("68adce01c0e1144d0a90295a") == "2025-08-26T15:08:49.000Z", "Failed Case 5"
print("PASS")
```

5. sleep_debt([8, 9, 10, 9, 10, 7], 7) should return 0.

```python
def sleep_debt(hours_slept, target_hours):
    ttl_tgt_hrs = (1 + len(hours_slept)) * target_hours
    ttl_zzz_hrs = sum(hours_slept)

    delta = 0 if ttl_zzz_hrs - ttl_tgt_hrs > 0 else ttl_tgt_hrs - ttl_zzz_hrs

    return delta


assert sleep_debt([6, 6, 6, 6, 6, 6], 8) == 20, "Failed Case 1"
assert sleep_debt([6, 7, 8, 4, 8, 6], 7) == 10, "Failed Case 2"
assert sleep_debt([10, 10, 9, 10, 9, 11], 9) == 4, "Failed Case 3"
assert sleep_debt([8, 7, 6, 7, 6, 8], 6) == 0, "Failed Case 4"
assert sleep_debt([8, 9, 10, 9, 10, 7], 7) == 0, "Failed Case 5"
print("PASS")
```

5. zip_strings("feCdCm", "reoeap") should return "freeCodeCamp".

Given two strings for different image dimensions, return the aspect ratio of the image with a greater width-to-height ratio.

    The given strings will be in the format "WxH", for example, "1920x1080".
    The aspect ratio is the ratio of width to height, reduced to the lowest whole numbers. For example, "1920x1080" reduces to "16:9".
    Return a string in format "W:H", for example, "16:9".

*Tests*
1. get_wider_aspect_ratio("1920x1080", "800x600") should return "16:9".
2. get_wider_aspect_ratio("1080x1350", "2048x1536") should return "4:3".

```python
from fractions import Fraction

def get_wider_aspect_ratio(a, b):
    fract_a = Fraction(a.replace('x', '/'))
    fract_b = Fraction(b.replace('x', '/'))
    wider = max(fract_a, fract_b)
    return f"{wider.numerator}:{wider.denominator}"


assert get_wider_aspect_ratio("1920x1080", "800x600") == "16:9"
assert get_wider_aspect_ratio("1080x1350", "2048x1536") == "4:3"
assert get_wider_aspect_ratio("640x480", "2440x1220") == "2:1"
assert get_wider_aspect_ratio("360x640", "1080x1920") == "9:16"
assert get_wider_aspect_ratio("3440x1440", "2048x858") == "43:18"
assert get_wider_aspect_ratio("12345x61234", "12534x51234") == "2089:8539"
print("PASS")
```
