**Day 1: Prank Number**

Given an array of numbers where all but one number follow a pattern, return a new array with the one number that doesn't follow the pattern fixed.

The pattern will be one of:
* The numbers increase from one to the next by a fixed amount (addition).
* The numbers decrease from one to the next by a fixed amount (subtraction).

For example, given [2, 4, 7, 8, 10] return [2, 4, 6, 8, 10].

*Tests*
1. fix_prank_number([2, 4, 7, 8, 10]) should return [2, 4, 6, 8, 10].
2. fix_prank_number([10, 10, 8, 7, 6]) should return [10, 9, 8, 7, 6].
3. fix_prank_number([12, 24, 36, 48, 61, 72, 84, 96]) should return [12, 24, 36, 48, 60, 72, 84, 96].
4. fix_prank_number([4, 1, -2, -5, -8, -5]) should return [4, 1, -2, -5, -8, -11].
5. fix_prank_number([0, 100, 200, 300, 150, 500]) should return [0, 100, 200, 300, 400, 500].
6. fix_prank_number([400, 425, 400, 375, 350, 325, 300]) should return [450, 425, 400, 375, 350, 325, 300].
7. fix_prank_number([-5, 5, 10, 15, 20]) should return [0, 5, 10, 15, 20].

```python
import numpy as np
from scipy import stats

def fix_prank_number(arr):
    diffs = np.diff(np.array(arr))

    # stats.mode(diffs): looks at all the np.diff() calculated jumps & finds the one that appears 
    # most often; e.g., [2, 4, 6, 10, 12] returns jumps: [2, 2, 4, 2] and mode == 2.
    #
    # keepdims=True: technical requirement for newer versions of Scipy to ensure the output shape stays consistent.
    # .mode[0]: Scipy returns the result as an array (in case there’s a tie). 
    # Adding [0] grabs the actual single number (e.g., 2) so you can use it for math.    
    expected_step = stats.mode(diffs, keepdims=True).mode[0]

    # diffs != expected_step: Tcreates a list of True/False values. 
    # It asks: "Is this jump different from our most common jump?"
    # e.g., [2, 2, 4, 2] != 2 becomes [False, False, True, False].
    # np.where(...): looks for the True values and tells you their position (the index)
    # .[0]: Like mode, np.where returns a tuple of arrays. 
    # The [0] extracts just the array of indices so you get a clean list like [2] instead of (array([2]),).
    gaps = np.where(diffs != expected_step)[0]

    if len(gaps) == 0: return arr

    gap = gaps[0]
    if len(gaps) == 1 and gap == 0:
        arr[0] = int(arr[1] - expected_step)
    else:
        arr[gap+1] = int(arr[gap] + expected_step)
    return arr

if (
    fix_prank_number([2, 4, 7, 8, 10]) == [2, 4, 6, 8, 10] and
    fix_prank_number([10, 10, 8, 7, 6]) == [10, 9, 8, 7, 6] and
    fix_prank_number([12, 24, 36, 48, 61, 72, 84, 96]) == [12, 24, 36, 48, 60, 72, 84, 96] and
    fix_prank_number([4, 1, -2, -5, -8, -5]) == [4, 1, -2, -5, -8, -11] and
    fix_prank_number([0, 100, 200, 300, 150, 500]) == [0, 100, 200, 300, 400, 500] and
    fix_prank_number([400, 425, 400, 375, 350, 325, 300]) == [450, 425, 400, 375, 350, 325, 300] and
    fix_prank_number([-5, 5, 10, 15, 20]) == [0, 5, 10, 15, 20]
):
    print('PASS')
else:
    print('FAIL')
```

```python
from collections import Counter

def fix_prank_number(arr):
    if len(arr) < 2: return arr
    
    # 1. np.diff() replacement: List Comprehension
    diffs = [arr[i+1] - arr[i] for i in range(len(arr)-1)]
    
    # 2. stats.mode() replacement: Counter
    # most_common(1) returns [ (value, count) ]
    expected_step = Counter(diffs).most_common(1)[0][0]
    
    # 3. np.where() replacement: Find indices of errors
    gaps = [i for i, d in enumerate(diffs) if d != expected_step]
    
    if not gaps:
        return arr
        
    first_gap = gaps[0]
    
    if len(gaps) == 1 and first_gap == 0:
        # Fix the very first number
        arr[0] = arr[1] - expected_step
    else:
        # Fix middle and end cases
        arr[first_gap + 1] = arr[first_gap] + expected_step
        
    return arr

if (
    fix_prank_number([2, 4, 7, 8, 10]) == [2, 4, 6, 8, 10] and
    fix_prank_number([10, 10, 8, 7, 6]) == [10, 9, 8, 7, 6] and
    fix_prank_number([12, 24, 36, 48, 61, 72, 84, 96]) == [12, 24, 36, 48, 60, 72, 84, 96] and
    fix_prank_number([4, 1, -2, -5, -8, -5]) == [4, 1, -2, -5, -8, -11] and
    fix_prank_number([0, 100, 200, 300, 150, 500]) == [0, 100, 200, 300, 400, 500] and
    fix_prank_number([400, 425, 400, 375, 350, 325, 300]) == [450, 425, 400, 375, 350, 325, 300] and
    fix_prank_number([-5, 5, 10, 15, 20]) == [0, 5, 10, 15, 20]
):
    print('PASS')
else:
    print('FAIL')
```

**Day 13: Name Initials**

Given a full name as a string, return their initials.
* Names to initialize are separated by a space.
* Initials should be made uppercase.
* Initials should be separated by dots.

For example, "Tommy Millwood" returns "T.M.".

*Tests*
1. get_initials("Tommy Millwood") should return "T.M.".
2. get_initials("Savanna Puddlesplash") should return "S.P.".
3. get_initials("Frances Cowell Conrad") should return "F.C.C.".
4. get_initials("Dragon") should return "D.".
5. get_initials("Dorothy Vera Clump Haverstock Norris") should return "D.V.C.H.N.".

```python
def get_initials(name):
    initials = []
    name_parts = name.split()
    for part in name_parts:
        # you cannot chain .append()
        initials.append(part[0])
        initials.append('.')
        # or simply
        # initials.append(part[0] + '.')
    return ''.join(initials)

if (
    get_initials("Tommy Millwood") == "T.M." and
    get_initials("Savanna Puddlesplash") == "S.P." and
    get_initials("Frances Cowell Conrad") == "F.C.C." and
    get_initials("Dragon") == "D." and
    get_initials("Dorothy Vera Clump Haverstock Norris") == "D.V.C.H.N."
):
    print('PASS')
else:
    print('FAIL')
```

**Day 14: Last Letter**

Given a string, return the letter from the string that appears last in the alphabet.
* If two or more letters tie for the last in the alphabet, return the first one.
* Ignore all non-letter characters.

*Tests*
1. get_last_letter("world") should return "w".
2. get_last_letter("Hello World") should return "W".
3. get_last_letter("The quick brown fox jumped over the lazy dog.") should return "z".
4. get_last_letter("HeLl0") should return "L".
5. get_last_letter("!#$ er@R asd fT.,> 2t0e9") should return "T".

```python
def get_last_letter(s):
    char = 'A'
    for letter in s:
        if letter.isalpha() and letter.upper() > char.upper():
            char = letter
    return char

if (
    get_last_letter("world") == "w" and
    get_last_letter("Hello World") == "W" and
    get_last_letter("The quick brown fox jumped over the lazy dog.") == "z" and
    get_last_letter("HeLl0") == "L" and
    get_last_letter("!#$ er@R asd fT.,> 2t0e9") == "T"
):
    print('PASS')
else:
    print('FAIL')
```

**Day 21: Odd Word**

Given a string of words, return only the words with an odd number of letters.
* Words in the given string will be separated by a single space.
* Return the words separated by a single space.

*Tests*
1. get_odd_words("This is a super good test") should return "a super".
2. get_odd_words("one two three four") should return "one two three".
3. get_odd_words("banana split sundae with rainbow sprinkles on top") should return "split rainbow sprinkles top".
4. get_odd_words("The quick brown fox jumped over the lazy river") should return "The quick brown fox the river"

```python
def get_odd_words(s):
    odd_words = []
    words = s.split()
    for word in words:
        if len(word) % 2 != 0:
            odd_words.append(word)
    return " ".join(odd_words)

if (
    get_odd_words("This is a super good test") == "a super" and
    get_odd_words("one two three four") == "one two three" and
    get_odd_words("banana split sundae with rainbow sprinkles on top") == "split rainbow sprinkles top" and
    get_odd_words("The quick brown fox jumped over the lazy river") == "The quick brown fox the river"
):
    print('PASS')
else:
    print('FAIL')
```

**Day 27: Word Score**

Given a word, return its score using a standard letter-value table:
| Letter | Value |
|   A 	 |   1   |
|   B    |   2   |
|  ...   | 	...  |
|   Z 	 |  26   |

* Upper and lowercase letters have the same value.

*Tests*
1. get_word_score("hi") should return 17.
2. get_word_score("hello") should return 52.
3. get_word_score("hippopotamus") should return 169.
4. get_word_score("freeCodeCamp") should return 94.

```python
def get_word_score(word):
    return sum(ord(letter.lower())-96 for letter in word)

if (
    get_word_score("hi") == 17 and
    get_word_score("hello") == 52 and
    get_word_score("hippopotamus") == 169 and
    get_word_score("freeCodeCamp") == 94    
):
    print('PASS')
else:
    print('FAIL')
```

**Day 29: URL Query Parser**

Given a URL that contains a query string, parse the query string into an object (or dictionary) of key-value pairs.

* The query string begins after the "?",
* each parameter is separated by "&",
* each key/value pair is separated by "="

For example, given "https://example.com/search?name=Alice&age=30", return:
{
  "name": "Alice",
  "age": "30"
}
All values should be returned as strings.

*Tests*
1. parse_url_query("https://example.com/search?name=Alice&age=30") should return {"name": "Alice", "age": "30"}
2. parse_url_query("https://freecodecamp.org/learn?skill=programming&language=python") should return {"skill": "programming", "language": "python"}
3. parse_url_query("https://freecodecamp.org/items?category=books&sort=asc&page=2") should return {"category": "books", "sort": "asc", "page": "2"}
4. parse_url_query("https://example.com?redirect=freecodecamp.org/learn&when=now") should return {"redirect": "freecodecamp.org/learn", "when": "now"}

```python
from urllib.parse import urlparse, parse_qsl

def parse_url_query(url):
    return dict(parse_qsl(urlparse(url).query))

if (
    parse_url_query("https://example.com/search?name=Alice&age=30") == {"name": "Alice", "age": "30"} and
    parse_url_query("https://freecodecamp.org/learn?skill=programming&language=python") == {"skill": "programming", "language": "python"} and
    parse_url_query("https://freecodecamp.org/items?category=books&sort=asc&page=2") == {"category": "books", "sort": "asc", "page": "2"} and
    parse_url_query("https://example.com?redirect=freecodecamp.org/learn&when=now") == {"redirect": "freecodecamp.org/learn", "when": "now"}    
):
    print('PASS')
else:
    print('FAIL')
```
